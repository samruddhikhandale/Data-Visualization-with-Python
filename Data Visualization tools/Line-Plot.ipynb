```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',sheet_name='Canada by Citizenship',skiprows=range(20),skipfooter=2)

df.drop(['AREA','REG','DEV','Type','Coverage'],axis=1,inplace=True)
df.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'},inplace=True)

years = list(map(int,range(1980,2014)))

df.set_index('Country',inplace=True)

indian_plot = df.loc['India',years]
indian_plot.index = indian_plot.index.map(int)
indian_plot.plot(kind='line')

plt.title('Immigration of Indians')
plt.ylabel('Immigrants')
plt.xlabel('Years')

plt.text(2005,35000,)
plt.show()
```
